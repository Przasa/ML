```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,1000)

# y=np.e**(x)
y_sigma=(1 / (1 + np.exp(-x)))

plt.plot(x,y_sigma)
plt.grid(True)
plt.title('g(x)=1/1+e^-x')
plt.xlabel('Prawdopodobienstwo: z=theta_0+theta_1*x_1...theta_n*x_n')
print(np.e)
```

```python
plt.figure(figsize=(8,6))
y_logA=-np.log(y_sigma)
y_logB=-np.log(1-y_sigma)
plt.plot(y_sigma,y_logA,label='Positives:\nCost(hsig(x),y)=-ln(hsig(x));  y=1',color='green')
# plt.set_title('Cost(hsig(x),y)=-ln(hsig(x));  y=1')
plt.plot(y_sigma,y_logB,label='Negatives:\nCost(hsig(x),y)=-ln(1-hsig(x)); y=-1',color='red')
# plt.set_title('Cost(hsig(x),y)=-ln(1-hsig(x)); y=-1')
plt.legend()
plt.grid(True)
# plt.axis([-1,1,-1,2])
```

```python
_x= np.linspace(0.01,1,100)
_y=np.log(_x)
_y1=-np.log(_x) #probki pozytywne
_y2=-np.log(-_x)
_y3=-np.log(1-_x) # probki negatywne

plt.plot(_x,_y,label='y=ln(p)')
plt.plot(_x,_y1,label='y=-ln(p)')
plt.plot(_x,_y2,label='y=-ln(-p)')
plt.plot(_x,_y3,label='y=-ln(1-p)')
plt.legend()
plt.grid(True)
```
